```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Binomial distribution parameters
p = .70 # probability of picking a champ
q = 1 - p # probability of not
n_trials = 10000

#Kelly params
wealth = 0
b = 1
f = p-q/b

trials = np.random.binomial(1, p, n_trials)
wager_size = 100
potential_net_payouts = [
    -31.57894737,
    -42.10526316,
    -31.57894737,
    36.84210526,
    -31.57894737,
    -55.26315789,
    -21.05263158,
    52.63157895,
    36.84210526,
    52.63157895,
    36.84210526,
    10.52631579,
    78.94736842,
    115.7894737,
    115.7894737,
    78.94736842,
    168.4210526,
    252.6315789,
    78.94736842,
]
```

```python
for j in range(50):
    wealth = 0
    record_wealth = []
    for i in trials:
        if i==1: # a player you picked won
            wealth = wealth + random.choice(potential_net_payouts)
        else:
            wealth = wealth - wager_size
        record_wealth.append(wealth)
    plt.plot(record_wealth)
plt.show()
    
```
